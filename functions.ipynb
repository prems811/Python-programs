```python
# calling the function with arbitrary argument
def sumOfNumbers(*num):
    res = 0

    for n in num:
        res = res + n
    print('Sum of all numbers = ', res)

sumOfNumbers(12,3)
sumOfNumbers(12,3,4)
sumOfNumbers(12,3,6)    


```

```python
# Global variable
discount = 10   # 10% discount

# Function 1: Display product details
def show_product_details(id, prodName, price):
    print("Product ID:", id)
    print("Product Name:", prodName)
    print("Price:", price)

    # Using global discount
    total_after_discount = price - (price * discount / 100)
    print("Price after", discount, "% discount:", total_after_discount)


# Function 2: Calculate total price using quantity
def calculate_total(price, quantity):
    # Local variable
    total_price = price * quantity
    
    # Apply global discount
    final_price = total_price - (total_price * discount / 100)
    
    return final_price


# now I am Calling functions 
show_product_details(101, "Laptop", 50000)

result = calculate_total(50000, 2)
print("Final total price for quantity:", result)
```

```python
# factorial of a function
def fact():
    x=5
    result = 1
    for i in range(1,x+1):
        result = result * i
        x= x-1
    print('factorial of 5 is = ', result)

fact()
```

```python
# Recursion example
def print_numbers(n):
    if n == 0:
        return 
    print_numbers(n - 1)
    print(n)

print_numbers(5)
```
