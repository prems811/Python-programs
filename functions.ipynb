```python
# calling the function with arbitrary argument
def sumOfNumbers(*num):
    res = 0

    for n in num:
        res = res + n
    print('Sum of all numbers = ', res)

sumOfNumbers(12,3)
sumOfNumbers(12,3,4)
sumOfNumbers(12,3,6)    

```
