```python
# if-else condition
age = 18
if age >= 18:
    print("can vote")
else:
    print('cannot vote')
```

```python
#if condition
countryName = "India"
if(countryName == "India"):
    print("Country name is correct")
```

```python
# Multiple if else condition
number = 100
if number > 500 :
    print('Number is greater than 50')
elif number > 300 :
    print('Number is greater than 300')
elif number > 400 :
    print('Number is greater than 400')
else:
    print('Number is less than 100')
    
```
