```python
#for loop
countryNames = ["India", "Austria", "USA", "China", "Pakistan"]
for c in countryNames:
    print(c ,end=", ")
```

```python
# For loop with range
for n in range(1,10):
    print(n)
```

```python
#nested for loop
countryNames = ["India", "Austria", "USA", "China", "Pakistan"]
number = [1,2,3,4,5]
for c in countryNames:
    for n in number:
        print(c," ", n)
```

```python
# While loop
number = 1

while number <= 10:
    print(number, end=", ")
    number = number + 1
```

```python
# Table of 5 using for for loop
number = 10
for n in range(1,11):
    print((n * 5) , end=", ")
```

```python
# Nested for loop
number = 5
for i in range(1,6):
    for i in range(1,6):
        print( i * number , end=" ,")
    number = number +1
    print('\n')
    
```

```python
# loop with break condition
for n in range(1,11):
    if(n==5):
        break
    print(n , end = " ")
```

```python
# loop with continue condition
for n in range(1,11):
    if(n==5):
        continue
    print(n , end = " ")
```

```python
# loop with pass keyword
for n in range(1,11):
    if(n==5):
        continue
    print(n , end = " ")
```
