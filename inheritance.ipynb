```python
# Single inheritance
class animal:
    name = ""
    def eat(self):
        print('This is my eat method inside ANIMAL class')


class dog (animal):
    def display(self):
        print(f"This is my display method inside DOG class")

objDog = dog()
objDog.eat()
objDog.display()
```
