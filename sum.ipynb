```python
class CalculateInterest:
    def simpleInterest(self, principle, rate, time):
        return principle * rate * time


class AxisBank(CalculateInterest):
    def simpleInterest(self, principle, rate, time):
        rate = rate * 10
        return super().simpleInterest(principle, rate, time)  


class KotakBank(CalculateInterest):
    def simpleInterest(self, principle, rate, time):
        rate = rate * 20
        return super().simpleInterest(principle, rate, time)


b = AxisBank()
print("Axis ROI:", b.simpleInterest(1000, 2, 1))

k = KotakBank()
print("Kotak ROI:", k.simpleInterest(1000, 2, 1))
```

```python
#single inheritance in python
class Bird:
    def flying(self):
        print("Bird is flying")

class parrot(Bird):
    def talking(self):
        print("Parrot is talking")

objParrot = parrot()
objParrot.flying()
objParrot.talking()
```

```python
# multi level inheritance in python.
class A:
    pass

class B(A):
    def printer(self):
        print("Printer method of class B is calling")

class C(B):
    def printer(self):
        print("Printer method of class C is calling")

objC = C()
objC.printer()
```

```python
#MEthod overriding in python
class A:
    pass

class B(A):
    def printer(self):
        print("Printer method of class B is calling")

class C(B):
    
    def printer(self):
        return super().printer()  #calling parent class method
        #print("Printer method of class C is calling") 

objC = C()
objC.printer()
```

```python
class Employee:
    def work(self): print("Employee working")

class Manager(Employee):
    def approve(self): print("Manager approved transaction")

class Admin(Manager):
    def report(self): print("Admin generated report")

emp = Employee(); 
emp.work()
mgr = Manager(); 
mgr.approve()
adm = Admin(); 
adm.report()
```
